A continuación se proponen una serie de ejercicios con funciones para que practiques. Pero recordemos primero cómo se define una función y cómo se utiliza. En el siguiente ejemplo, 'cuadrado' es el nombre de mi función, 'x' es el argumento y como hay un 'return' sabremos que devuelve un valor a la salida cuando usamos dicha función.

```python
def cuadrado(x):
  return x**2

y = 4
y2 = cuadrado(y)
print('El cuadrado de', y, 'es', y2)
```

Lo interesante de las funciones es que puedo usarlas cuantas veces quiera en mi programa una vez que estén definidas.

```python
cuadrado(11)
```

**None** es un tipo de datos especial que significa valor nulo o vacío. Es muy útil en el preprocesado de datos o cuando queremos devolver en una función un error.

```python
cosa = None
if cosa is None:
    print("Es nada")
else:
    print("Es algo")
```

> **Ejercicio**: Diseñe una función llamada *area_circulo* que reciba como parámetro el valor de un radio y devuelva el área de un círculo con dicho radio.

```python
from math import pi


def area_circulo(radio):
  area = pi * radio ** 2 
  return area


R = float(input('Introduce el radio:'))
miArea = area_circulo(R)
print('El area es', miArea)

miArea2 = area_circulo(90)
print('El area es', miArea2)
```

> **Ejercicio**: Diseñe una función sin argumentos que devuelva el valor de un número aleatorio entre 0 y 10. Puedes usar la función *random()*.

```python
#<SOL>
from random import random

def devuelveAleatorio():
  num_aleatorio = 10 * random()
  return num_aleatorio

print('Numero aleatorio: ', devuelveAleatorio())

#</SOL>
```

```python
#<SOL>
from random import random, randint

def aleatorioEntero():
  return randint(0, 10)

def aleatorioFloat():
  return 10*random()    # Pasamos del rango [0, 1] -----> [0, 10]

print('Numero aleatorio entero: ', aleatorioEntero())
print('Numero aleatorio flotante: ', aleatorioFloat())
#</SOL>
```

### 1.4 Especificar parámetros con valores por defecto (Extra!)

Es posible especificar valores por defecto en los parámetros de una función. El valor por defecto se usará cuando en la llamada a la función no se haya proporcionado el correspondiente argumento.

```python
def menu(bebida, entrante, postre='Tiramisú'):
    return {'bebida': bebida, 'entrante': entrante, 'postre': postre}
```

```python
# Hacemos uso del valor por defecto del parámetro "postre"
menu('Vino', 'Ensalada')
```

```python
# "Sobreescribimos" el valor de "postre" especificando uno concreto
menu('Agua', 'Revuelto de setas', 'Helado')
```

Los valores por defecto en los parámetros se calculan cuando se **define** la función, no cuando se **ejecuta**.
